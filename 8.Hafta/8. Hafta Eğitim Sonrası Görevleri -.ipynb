# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
def vki_hesapla(kilo, boy):
    return kilo / (boy ** 2)

def vki_siniflandir(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki <= 24.9:
        return "Normal Kilolu"
    elif 25 <= vki <= 29.9:
        return "Fazla Kilolu"
    elif 30 <= vki <= 39.9:
        return "Obez"
    else:
        return "İleri Derecede Obez (Morbid Obez)"

def veri_girisi():
    try:
        kac_giris = int(input("Kaç kişilik veri gireceksiniz? "))
    except ValueError:
        print("Lütfen geçerli bir sayı giriniz.")
        return

    with open("vki_verileri.txt", "w") as dosya:
        dosya.write("İsim Soyisim, Yaş, Boy, Kilo, VKİ, Sınıflandırma\n")
        
        for i in range(kac_giris):
            print(f"\n{i+1}. kişi için bilgi giriniz:")
            isim = input("İsim: ")
            soyisim = input("Soyisim: ")
            try:
                yas = int(input("Yaş: "))
                boy = float(input("Boy (cm cinsinden): ")) / 100  
                kilo = float(input("Kilo (kg cinsinden): "))
            except ValueError:
                print("Lütfen geçerli bir değer giriniz.")
                continue

            vki = vki_hesapla(kilo, boy)
            sinif = vki_siniflandir(vki)

            dosya.write(f"{isim} {soyisim}, {yas}, {boy:.2f}, {kilo}, {vki:.2f}, {sinif}\n")
            print(f"\nVKİ: {vki:.2f}, Sınıflandırma: {sinif}")
            
    print("\nVeriler başarıyla kaydedildi.")

veri_girisi()

```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import datetime
import json

menu = {
    "Çilek": 5.0,
    "Muz": 3.0,
    "Yaban Mersini": 4.5,
    "Ispanak": 2.5,
    "Badem Sütü": 3.5,
    "Yulaf": 2.0
}

def menu_goster():
    print("\nYetBar Menü:")
    for urun, fiyat in menu.items():
        print(f"{urun}: {fiyat} TL")
    print("\n")

def satin_al():
    toplam_fiyat = 0
    secimler = []

    while True:
        urun = input("Smoothie'nize eklemek istediğiniz ürünü seçin (Çıkmak için 'q' tuşuna basın): ").title()
        
        if urun == 'Q':
            break
        elif urun in menu:
            secimler.append(urun)
            toplam_fiyat += menu[urun]
            print(f"{urun} eklendi. Toplam Fiyat: {toplam_fiyat} TL")
        else:
            print("Geçersiz ürün. Lütfen menüden bir ürün seçin.")
    
    if toplam_fiyat > 0:
        tarih = datetime.datetime.now()
        satis_bilgisi = {
            "tarih": tarih.strftime("%Y-%m-%d %H:%M:%S"),
            "urunler": secimler,
            "toplam_fiyat": toplam_fiyat
        }
        kaydet_json(satis_bilgisi)
        print(f"\nSatın alma işlemi tamamlandı. Toplam: {toplam_fiyat} TL")
    else:
        print("\nHiçbir ürün seçmediniz.")

def kaydet_json(satis_bilgisi):
    try:
        with open("satis_kaydi.json", "r") as dosya:
            satislar = json.load(dosya)
    except FileNotFoundError:
        satislar = []

    satislar.append(satis_bilgisi)
    
    with open("satis_kaydi.json", "w") as dosya:
        json.dump(satislar, dosya, indent=4)

def yetbar():
    print("YetBar - Smoothie Bank'a hoş geldiniz!")

    while True:
        print("\n1. Menüyü Görüntüle")
        print("2. Satın Alma")
        print("3. Çıkış")
        secim = input("Bir seçenek girin (1, 2, 3): ")

        if secim == '1':
            menu_goster()
        elif secim == '2':
            satin_al()
            devam = input("\nBaşka bir işlem yapmak ister misiniz? (E/H): ").lower()
            if devam != 'e':
                break
        elif secim == '3':
            print("Çıkış yapılıyor. YetBar'ı tercih ettiğiniz için teşekkür ederiz!")
            break
        else:
            print("Geçersiz seçenek. Lütfen 1, 2 veya 3 girin.")

yetbar()
```
