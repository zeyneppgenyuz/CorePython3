````python
# 1- "Bmw, Mercedes, Opel, Mazda" elemanlarına sahip bir liste oluşturunuz.
arabalar = ["Bmw", "Mercedes", "Opel", "Mazda"]

# 2- Liste Kaç elemanlıdır ?
print(len(arabalar))  # Çıktı: 4

# 3- Listenin ilk ve son elemanı nedir ?
print(arabalar[0])    # Çıktı: Bmw
print(arabalar[-1])   # Çıktı: Mazda

# 4- Mazda değerini Toyota ile değiştirin.
arabalar[-1] = "Toyota"
print(arabalar)  # Çıktı: ['Bmw', 'Mercedes', 'Opel', 'Toyota']

# 5- Listenin -2 indeksindeki değer nedir ?
print(arabalar[-2])  # Çıktı: Opel

# 6- Listenin ilk 3 elemanını alın.
print(arabalar[:3])  # Çıktı: ['Bmw', 'Mercedes', 'Opel']

# 7- Listenin son 2 elemanı yerine "Toyota" ve "Renault" değerlerini ekleyin.
arabalar[-2:] = ["Toyota", "Renault"]
print(arabalar)  # Çıktı: ['Bmw', 'Mercedes', 'Toyota', 'Renault']

# 8- Listenin üzerine "Audi" ve "Nissan" değerlerini ekleyin.
arabalar += ["Audi", "Nissan"]
print(arabalar)  # Çıktı: ['Bmw', 'Mercedes', 'Toyota', 'Renault', 'Audi', 'Nissan']

# 9- Listenin son elemanını silin.
arabalar.pop()
print(arabalar)  # Çıktı: ['Bmw', 'Mercedes', 'Toyota', 'Renault', 'Audi']

# 10- Liste elemanlarını tersten yazdırınız.
print(arabalar[::-1])  # Çıktı: ['Audi', 'Renault', 'Toyota', 'Mercedes', 'Bmw']

# 11- Aşağıdaki verileri bir liste içinde saklayınız.
students = [
    {'name': 'Yiğit Bilgi', 'year': 2010, 'grades': (70, 60, 70)},
    {'name': 'Sena Turan', 'year': 1999, 'grades': (80, 80, 70)},
    {'name': 'Ahmet Turan', 'year': 1998, 'grades': (80, 70, 90)}
]

# 12- Liste elemanlarını ekrana yazdırınız.
for student in students:
    print(student)

names = ['Ali', 'Yağmur', 'Hakan', 'Deniz']
years = [1998, 2000, 1998, 1987]

# 13- "Cenk" ismini listenin sonuna ekleyiniz.
names.append("Cenk")
print(names)  # Çıktı: ['Ali', 'Yağmur', 'Hakan', 'Deniz', 'Cenk']

# 14- "Sena" değerini listenin başına ekleyiniz.
names.insert(0, "Sena")
print(names)  # Çıktı: ['Sena', 'Ali', 'Yağmur', 'Hakan', 'Deniz', 'Cenk']

# 15- "Deniz" ismini listeden siliniz.
names.remove("Deniz")
print(names)  # Çıktı: ['Sena', 'Ali', 'Yağmur', 'Hakan', 'Cenk']

# 16- "Ali" listenin bir elemanı mıdır?
print("Ali" in names)  # Çıktı: True

# 17- Liste elemanlarını ters çevirin.
names.reverse()
print(names)  # Çıktı: ['Cenk', 'Hakan', 'Yağmur', 'Ali', 'Sena']

# 18- Liste elemanlarını alfabetik olarak sıralayınız.
names.sort()
print(names)  # Çıktı: ['Ali', 'Cenk', 'Hakan', 'Sena', 'Yağmur']

# 19- years listesini rakamsal büyüklüğe göre sıralayınız.
years.sort()
print(years)  # Çıktı: [1987, 1998, 1998, 2000]

# 20- str = "Chevrolet,Dacia" karakter dizisini listeye çeviriniz.
str = "Chevrolet,Dacia"
car_list = str.split(',')
print(car_list)  # Çıktı: ['Chevrolet', 'Dacia']

# 21- years dizisinde kaç tane 1998 değeri vardır?
count_1998 = years.count(1998)
print(count_1998)  # Çıktı: 2

# 22- years dizisinin tüm elemanlarını siliniz.
years.clear()
print(years)  # Çıktı: []

# 23- Kullanıcıdan alacağınız 3 tane marka bilgisini bir listede saklayınız.
brands = []
for _ in range(3):
    brand = input("Bir marka girin: ")
    brands.append(brand)
print(brands)
````
